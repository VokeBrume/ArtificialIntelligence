```python
# FUNCTIONS FOR ANALYSIS

# Function to split tabs of excel file into multiple files
# @param: 
# @return: 
def split_excel_wooksheets(file_path, file_name):
  excel_file = pd.ExcelFile(file_path + file_name)
  for sheet in excel_file.sheet_names:
    df = pd.read_excel(excel_file,sheet_name=sheet)
    df.to_excel(f"{file_path}{sheet}.xlsx",index=False)
```
